```python
import pandas as pd

# Load the CSV file for Tanzania
df = pd.read_csv(r"D:\Aman\KAIM-9\Week-0\Week0-climate-challange\climate-challenge-week0\data\tanzania.csv")

# Add a Country column
df["Country"] = "Tanzania"

# Convert YEAR and DOY to a proper datetime column
df["Date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")

# Extract Month as a separate column
df["Month"] = df["Date"].dt.month

df.head()
```

```python
import numpy as np

# Replace all -999 with np.nan
df = df.replace(-999, np.nan)

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. New shape: {df.shape}")
```
