```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import seaborn as sns

dfs = []

countries = ["ethiopia", "kenya", "nigeria", "sudan", "tanzania"]

for country in countries:
    df = pd.read_csv(f"D:\Aman\KAIM-9\Week-0\Week0-climate-challange\climate-challenge-week0\data\{country}_clean.csv")
    dfs.append(df)

combined = pd.concat(dfs)
```

## Temprature comparision

```python
combined.tail()
combined.to_csv(r"D:\Aman\KAIM-9\Week-0\Week0-climate-challange\climate-challenge-week0\data\cleaned_all.csv", index=False)
```

```python
sns.lineplot(
    data=combined,
    x="Date",
    y="T2M",
    hue="Country"
)
```

## Rainfall variability

```python
sns.boxplot(
    data=combined,
    x="Country",
    y="PRECTOTCORR"
)
```

## Extreme heat days

```python
extreme_heat = combined[
    combined["T2M_MAX"] > 35
].groupby("Country").size()
```

## Dry Days

```python
dry_days = combined[
    combined["PRECTOTCORR"] < 1
].groupby("Country").size()
```

## Statistical testing

```python
from scipy.stats import f_oneway

ethiopia = combined[combined["Country"]=="Ethiopia"]["T2M"]
kenya = combined[combined["Country"]=="Kenya"]["T2M"]
nigeria = combined[combined["Country"]=="Nigeria"]["T2M"]

f_oneway(ethiopia, kenya, nigeria)
```
